```python
import matplotlib.pyplot as plt
import numpy as np

# Erstellen von Daten
x = np.linspace(0, 11, 100)  # Erzeugt 100 Punkte zwischen 0 und 10
y = np.sin(x)               # Berechnet den Sinus jedes x-Wertes

# Erstellen des Plots
plt.figure(figsize=(10, 6))  # Festlegen der Größe des Plots
plt.plot(x, y, label='sin(x)')  # Hinzufügen der Daten zum Plot
plt.title('Beispielplot eines Liniendiagramms')  # Titel des Plots
plt.xlabel('x-Werte')  # Beschriftung der x-Achse
plt.ylabel('y-Werte')  # Beschriftung der y-Achse
plt.legend()          # Zeigt die Legende an
plt.grid(True)        # Fügt ein Gitter hinzu
plt.show()
```
